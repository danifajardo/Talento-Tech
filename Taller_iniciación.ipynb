- Ejercicio 1: Mostrar las primeras filas del dataframe (df_ventas)

```python
print(df_ventas.head(5))
```

- Ejercicio 2: Calcular el total de ventas por región

```python
ventas_por_region = df_ventas.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
```

- Ejercicio 3: Filtrar el data frame sólo para mostrar la ventas hechas por David

```python
ventas_david = df_ventas[df_ventas['Vendedor'] == 'David']
print(ventas_david)
```

- Ejercicio 4: Calcular el promedio de ventas por región

```python
#mi forma
promedio_ventas_region = df_ventas.groupby('Region')['Ventas'].sum()//df_ventas.groupby('Region')['Ventas'].count()
print(promedio_ventas_region)
print('---------------------------')
#forma del profe
promedio_ventas_region = df_ventas.groupby('Region')['Ventas'].mean()
print(promedio_ventas_region)
```

- Ejercicio 5: Calcular el promedio de ventas por región con sólo dos decimales

```python
promedio_ventas_region = df_ventas.groupby('Region')['Ventas'].mean().round(2)
print(promedio_ventas_region)
```

- Ejercicio 6: Ordenar e data-frema por ventas en orden descendante

```python
df_ordenado = df_ventas.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
```

- Ejercicio 7: Filtrar las filas con ventas mayores a 10000

```python
df_filtrado = df_ventas[df_ventas['Ventas'] > 10000].sort_values(by='Ventas', ascending=True)
print(df_filtrado)
```

- Ejercicio 8: Eliminar columna vendedor

```python
df_sin_vendedor = df_ventas.drop(columns=['Vendedor'])
print(df_sin_vendedor)
```
